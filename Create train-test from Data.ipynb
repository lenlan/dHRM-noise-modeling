# Import and adapt uMelts

```python
df_58_norm_shifted = pd.read_csv('df_58_norm_shifted.csv', index_col=0)
df_58_norm_shifted
```

```python
chip1_dfs = [pd.read_csv('Chips1/'+str(i)+'.csv').iloc[:,1:] for i in range(10)]
chip1_residuals = [df - df.mean(axis=0).values for df in chip1_dfs]
```

# Add noise to UMelts

```python
all_all=[]
for j in range(3):
    all_perturbed = []
    np.random.seed(j)
    for i in range(10):
        model_nr = i
        model = chip1_residuals[model_nr]
        residuals_model_padded = pd.DataFrame([padding_function3(i) for i in model.values]) 
        indices = np.random.choice(len(model), len(model)//2)
        residuals_dfs = []
        residuals_dfs.append(residuals_model_padded.iloc[indices])
        residuals_dfs.append(residuals_model_padded.iloc[~indices])
        mean = chip1_dfs[model_nr].mean(axis=0)
        peak_height = max(mean)-min(mean)
        peak_loc = float(mean.idxmax())
        print(peak_height, peak_loc)

        all_perturbed.append([perturb_shifted_delta(2, len(residuals_dfs[0]), df_58_norm_shifted.loc[duo], residuals_dfs, peak_loc, peak_height) for duo in euclid_duos])
    all_all.append(all_perturbed)
```

# Write Files

```python
import os

for l in range(10):
    os.mkdir('noise'+str(l))
    folder = 'noise'+str(l)+'/'
    for k in range(50):
        os.mkdir(folder+'duo'+str(k))
        for h in range(3):
            os.mkdir(folder+'duo'+str(k)+'/split'+str(h))
            n_train = int(len(all_all[h][l][k][0])*2/3)
            for j in range(3):
                x_train, x_test = ([] for i in range(2))  
                for i in range(2):
                    data = all_all[h][l][k][i].values
                    print(data.shape)
                    np.random.seed(j)
                    np.random.shuffle(data)
                    x_train.append(data[:n_train,:])
                    x_test.append(data[n_train:,:])

                x_train = pd.DataFrame(np.vstack(x_train))
                x_test = pd.DataFrame(np.vstack(x_test))
                x_train.to_csv(folder+'duo'+str(k)+'/split'+str(h)+'/x_train'+str(j)+'.csv')
                x_test.to_csv(folder+'duo'+str(k)+'/split'+str(h)+'/x_test'+str(j)+'.csv')
```

# Functions

```python
from scipy.interpolate import interp1d
def padding_function3(i):
    x = np.concatenate((np.arange(5),np.arange(15,20)))
    y = np.concatenate((i[-5:], i[:5]))
    f = interp1d(x, y, kind='cubic')
    return np.concatenate((f(np.arange(10,15)),i,f(np.arange(5,10))))

def perturb_shifted_delta(n_org, n_curves, df_umelt, residuals_dfs, peak_loc, peak_height):   
    perturbed_2 = []
    for i in range(n_org):
        perturbed = []
        mean = df_umelt.iloc[i].values
        index = df_umelt.iloc[i].index.values
        maxidx = float(df_umelt.iloc[i].idxmax())
        ratio = max(df_umelt.iloc[i].values)/peak_height
        
        for k in range(n_curves):
            perturb = mean+np.roll(residuals_dfs[i].iloc[k].values, int((maxidx-peak_loc)*10))*ratio
            perturbed.append(perturb)
        perturbed_2.append(pd.DataFrame(perturbed) )
    return perturbed_2
```
